### User Defined Function to Convert String to Json and add them as columns to the dataframe

```python
def break_fields(df, column_name, column_prefix):
    user_values = df[column_name]
    #print(user_values)
    user_values = json.loads(json.dumps(ast.literal_eval(user_values)))
    for key, value in user_values.items():
        key = column_prefix + key
        df[key] = value
    return df
        
eg = nov_2

eg = eg.apply(lambda x: break_fields(x, 'user', 'user_'), axis=1)
eg = eg.apply(lambda x: break_fields(x, 'retweeted_status', 'retweet_'), axis=1)
```

### Cleaning the text

#### Removing Punctuations

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]', '', text)
    text=re.sub(r'[RT]+','',text)#replace RT-tags
    return text

nov_2_df['Tweet_punct'] = nov_2_df['full_text'].apply(lambda x: remove_punct(x))
nov_2_df.head(10)
```

#### Tokenization

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

nov_2_df['Tweet_tokenized'] = nov_2_df['Tweet_punct'].apply(lambda x: tokenization(x.lower()))
nov_2_df.head()
```

#### Remove stop words

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
nov_2_df['Tweet_nonstop'] = nov_2_df['Tweet_tokenized'].apply(lambda x: remove_stopwords(x))
nov_2_df.head(10)
```

#### Stemming and Lammitization

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

nov_2_df['Tweet_stemmed'] = nov_2_df['Tweet_nonstop'].apply(lambda x: stemming(x))
nov_2_df.head()
```

```python
nov_2_df['Tweet_stemmed'][0]
```

```python
nov_2_df['Tweet_stemmed'][1]
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

nov_2_df['Tweet_lemmatized'] = nov_2_df['Tweet_nonstop'].apply(lambda x: lemmatizer(x))
nov_2_df.head()
```

```python
nov_2_df['Tweet_lemmatized'][0]
```

```python
nov_2_df['Tweet_lemmatized'][1]
```

```python
#Calculating Negative, Positive, Neutral and Compound values
nov_2_df[['polarity','subjectivity']] = nov_2_df['Tweet_punct'].apply(lambda Text: pd.Series(TextBlob(''.join(Text)).sentiment))
for index, row in nov_2_df['Tweet_punct'].iteritems():
    score = SentimentIntensityAnalyzer().polarity_scores(''.join(row))
    neg = score['neg']
    neu = score['neu']
    pos = score['pos']
    comp = score['compound']
    if neg > pos:
        nov_2_df.loc[index, 'sentiment'] = "negative"
    elif pos > neg:
        nov_2_df.loc[index, 'sentiment'] = "positive"
    else:
        nov_2_df.loc[index, 'sentiment'] = "neutral"
        nov_2_df.loc[index, 'neg'] = neg
        nov_2_df.loc[index, 'neu'] = neu
        nov_2_df.loc[index, 'pos'] = pos
        nov_2_df.loc[index, 'compound'] = comp
```

```python
from gensim.corpora import Dictionary

#create dictionary
text_dict = Dictionary(nov_2_df['Tweet_lemmatized'])

#view integer mappings
text_dict.token2id
```

```python
tweets_bow = [text_dict.doc2bow(tweet) for tweet in nov_2_df['Tweet_lemmatized']]
```

```python
k = 5
tweets_lda = LdaModel(tweets_bow,
                      num_topics = k,
                      id2word = text_dict,
                      random_state = 1,
                      passes=10)

tweets_lda.show_topics()
```
